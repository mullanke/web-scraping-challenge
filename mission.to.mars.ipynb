```python
from bs4 import BeautifulSoup as bs
import requests
from splinter import Browser
import pandas as pd
from webdriver_manager.chrome import ChromeDriverManager
```

```python
#scrape Nasa's news page
url = "https://mars.nasa.gov/news/"
response = requests.get(url)

#create beautiful soup object

soup = bs(response.text, 'html.parser')
```

```python
#find only gets the first instance of content_title from the web scraping pull. This will be the most recent story at the time of retrieval

firsttitle = soup.find('div', class_='content_title').get_text()

print(firsttitle)
```

```python
#same method to get the first news story's description
firstdescription = soup.find('div', class_="rollover_description_inner").get_text()

print(firstdescription)
```

Mars Facts
-Visit the Mars Facts webpage here and use Pandas to scrape the table containing facts about the planet including Diameter, Mass, etc.
-Use Pandas to convert the data to a HTML table string.

```python
marsurl= "https://space-facts.com/mars/"

response = requests.get(marsurl)

soup= bs(response.text, 'html.parser')

# print(soup.prettify())
```

```python
#grab the table from the URL
table = pd.read_html(marsurl)
mars_df = table[0]
mars_df.columns = ['Statistics', "Mars Measurements"]

mars_df = pd.DataFrame(mars_df)
mars_df.head()
```

```python
#organize and clean up the table formatting
stats = pd.Series(mars_df['Statistics'])
mars_df['Statistics'] = stats.str.strip(':')
mars_df = mars_df.set_index('Statistics')
mars_df.head()
```

```python
#send to html
mars_df.to_html('marsfacts.html')
```

--Append the dictionary with the image url string and the hemisphere title to a list. This list will contain one dictionary for each hemisphere.

```python
from selenium import webdriver

link = "https://accounts.google.com"
driver = webdriver.Chrome(executable_path='/path/to/chromedriver')
driver.get(link)
```

```python
marshemi = "https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars"

response = requests.get(marshemi)

soup= bs(response.text, 'html.parser')

base_url = 'https://astrogeology.usgs.gov'

executable_path = {'executable_path': '/Users/kelliemullan/.wdm/drivers/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
items = soup.find_all('div', class_='item')

urls = []

titles = []

for x in items:
    urls.append(base_url + x.find('a')['href'])
    titles.append(x.find('h3').text.strip())
print(urls)
print(titles)
```

```python
#high res photos
executable_path = {'executable_path': '/Users/kelliemullan/.wdm/drivers/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
browser.visit(marshemi)
```

```python
browser.visit(urls[0])
html = browser.html
soup = bs(html, 'html.parser')
exampleurl = base_url+soup.find('img',class_='wide-image')['src']
exampleurl
```

```python
img_urls = []
for exampleurl in urls:
    browser.visit(exampleurl)
    html = browser.html
    soup = bs(html, 'html.parser')
    exampleurl = base_url+soup.find('img',class_='wide-image')['src']
    img_urls.append(exampleurl)
    
img_urls
```

```python
#titles that go with images

hemispheretitles = []

for x in range (len(titles)):
    hemispheretitles.append({'Title of Image': titles[x], 'URL': img_urls[x]})

for x in range(len(hemispheretitles)):
    print(hemispheretitles[x]['Title of Image'])
    print(hemispheretitles[x]['URL'] + '\n')
```
